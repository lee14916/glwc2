```python
def getDat3pt(tf,Gc,iso):
    i_Gc={'id':0,'gx':1,'gy':2,'gz':3,'gt':4,'g5':5,'g5gx':6,'g5gy':7,'g5gz':8,'g5gt':9}[Gc]
    sgn_iso={'+':+1,'-':-1}[iso]
    projA=projB=((0,0,0),'a')
    proj='P0'
    return data['NJN']['dt'+str(tf)+'_protonup'][projA][projB][proj]['MprotonUp'][:,1:tf,i_Gc]+sgn_iso*data['NJN']['dt'+str(tf)+'_protondn'][projA][projB][proj]['MprotonDn'][:,1:tf,i_Gc]
    

def getDat(Gc,iso):
    proj=((0,0,0),'a')
    i_Gc={'id':0,'gx':1,'gy':2,'gz':3,'gt':4,'g5':5,'g5gx':6,'g5gy':7,'g5gz':8,'g5gt':9}[Gc]
    str_iso={'+':'u+d','-':'u-d'}[iso]

    dat2pt=data['NN']['N_2pt'][proj][proj]['P0']['NP']
    dat3pt={}
    for tf in tfList:
        dat3pt[tf]=getDat3pt(tf,Gc,iso)

    dat3ptDISC={}
    for tf in tfList:
        dat3ptDISC[tf]=data['NJN']['N_IL'][proj][proj]['P0']['NP_'+str_iso+'_deltat_'+str(tf)][:,1:tf,i_Gc]

    datVEV=np.zeros(len(cfgList))
    if Gc=='id' and iso=='+':
        datVEV=data['others']['u+d_id_vev']
    if Gc=='g5' and iso=='-':
        datVEV=data['others']['u-d_g5_vev']

    t=np.array([
        np.array([dat2pt[cfg]]+[dat3pt[tf][cfg] for tf in tfList]+[dat3ptDISC[tf][cfg] for tf in tfList]+[datVEV[cfg]],dtype=object)
    for cfg in range(len(cfgList))], dtype=object)

    return t
```

```python
dat=getDat('id','+')
case=['conn','disc'][1]
preFactor=1

def tFunc(dat):
    t=np.mean(dat,axis=0)
    t2pt=t[0]
    t3pt=t[1:1+3]
    t3ptDISC=t[4:4+3]
    tVEV=t[7]
    if case == 'conn':
        t=np.array([np.real( preFactor*(t3pt[i_tf]/t2pt[tfList[i_tf]])) for i_tf in range(len(tfList))],dtype=object)
    elif case == 'disc':
        t=np.array([np.real( preFactor*(t3ptDISC[i_tf]/t2pt[tfList[i_tf]]-tVEV)) for i_tf in range(len(tfList))],dtype=object)
    return t

(mean,cov,err)=yu.jackknife(dat,tFunc,2)

(Nrow, Ncol) = (1, 1)
fig, axs = plt.subplots(Nrow, Ncol, figsize=(5*Ncol, 4*Nrow), squeeze=False)

irow=0;icol=0
for i_tf in range(len(tfList)):
    tf=tfList[i_tf]
    tMean=mean[i_tf];tErr=err[i_tf]
    axs[irow,icol].errorbar(np.arange(1 - tf//2,tf//2),tMean,tErr,\
        fmt='o',color=['r','g','b'][i_tf],elinewidth=2,capsize=6,mfc='white')
axs[irow,icol].set_xticks(range(-6,6,2))
None
```
